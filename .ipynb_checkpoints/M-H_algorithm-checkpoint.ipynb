```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import beta
%matplotlib inline
```

```python
plt.style.use("ggplot")
np.random.seed(123)
```

```python
# β分布のパラメータを設定
a, b = 1.5, 2.0
# β分布の範囲に合わせてxの範囲を設定
x = np.linspace(beta.ppf(0.001, a, b), beta.ppf(0.999, a, b), 100)

plt.plot(x, beta.pdf(x, a, b))
```

```python
# 初期値を決める
theta = 0.8

NMCS = 20000

epsilon = 0.5
# 本来はこのepsilonの値をうまく設定する必要がある。

theta_mcs = [theta]

for i in range(NMCS):
    # 平均0、標準偏差1の正規分布でランダムウォークさせる
    theta_new = theta + epsilon * np.random.randn()
    
    # f(θ')とf(θ)の大小判定
    # f(θ')の方が大きい場合
    if beta.pdf(theta_new, a, b) > beta.pdf(theta, a, b):
        # θ'→θの流れの方がθ→θ'より強いので受け入れる
        theta = theta_new
    # f(θ)の方が大きい場合
    else:
        # 補正係数を計算する
        # 補正係数はθ'とθの流れが近いほど大きくなる（１に近づく）
        r = beta.pdf(theta_new, a, b) / beta.pdf(theta, a, b)
        # 補正係数が乱数より大きい場合、受け入れる
        # θとθ'の流れが近いほど、受け入れやすくなる
        # θの方が圧倒的に大きいと、まだまだ受け入れずに進む必要がある
        if np.random.rand() < r:
            theta = theta_new
    
    theta_mcs.append(theta)
df = pd.DataFrame(theta_mcs)
```

```python
df.head()
```

```python
plt.plot(df[0])
plt.xlabel("MCS")
plt.ylabel("$\Theta$")
```

```python
# ヒストグラムを作成する
# 最初の1000ぐらいはバーンインとしてカウントしない
plt.hist(df[0][1000:], normed=True, bins=40)
x = np.linspace(beta.ppf(0.001, a, b), beta.ppf(0.999, a, b), 100)
plt.plot(x, beta.pdf(x, a, b))
```
