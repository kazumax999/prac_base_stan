```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use("ggplot")
np.random.seed(123)
```

```python
# 確率の行列のひな型を作る
p_trans = np.zeros([3,3])
```

```python
# 遷移確率の行列を作成する
# 0: office, 1: kyuukei, 2: jikken
p_trans[0,0] = 0.1
p_trans[0,1] = 0.2
p_trans[0,2] = 0.7
p_trans[1,0] = 0.1
p_trans[1,1] = 0.4
p_trans[1,2] = 0.5
p_trans[2,0] = 0.3
p_trans[2,1] = 0.3
p_trans[2,2] = 0.4
```

```python
p_trans
```

```python
NMCS = 400

# 初期状態の設定
c_state = 0

c_arr = [c_state]

for i in range(NMCS):
    # 0, 1, 2から今いる場所の遷移核に応じて数字を選択する
    current = np.random.choice(3, 1, p=p_trans[c_state, :])
    # choice(3, 1) 0,1,2からランダムに選ぶ。ここにp_transで重みづけをする
    # アレイで返ってくるので0番目を現在の状態に設定する
    c_state = current[0]
    # 現在の状態を状態の配列に追加する
    c_arr.append(c_state)

# 描画するためにDataFrameに追加
df = pd.DataFrame(c_arr)
plt.plot(df[0])
plt.xlabel("MCS")
plt.ylabel("place")
    
```

```python
# ヒストグラムを描画する
plt.hist([ df[0][:10], df[0][:50], df[0][:200], df[0][:400] ], normed=True, label= ["10", "50", "200", "400"])

plt.xlabel("place")
plt.ylabel("frequency")
plt.legend()
```
